```python
from pathlib import Path
from cloudpathlib import S3Path, S3Client
import nibabel as nib
import neuropythy as ny

# Make sure that we have a cache path:
cache_path = Path('/tmp/cache')
if not cache_path.exists():
    cache_path.mkdir()

#tell the S3Path and S3Client object that our credentials are labeled 'hcp'
hcp_base_path = S3Path(
    's3://hcp-openaccess/HCP_1200/',
    client=S3Client(
        local_cache_dir=cache_path,
        profile_name='hcp'))
```

```python
#HCP uses 6-digit subject IDs 
#list the contents of the base path 

hcp_subdirs = list(hcp_base_path.iterdir())
# There are about 1200 of these, so just show the first 10:
hcp_subdirs[:10]
```

```python
int(str(hcp_subdirs[0])[-6:])
```

```python
#just testing
path_string = hcp_subdirs[0] / 'MNINonLinear' / 'Results' / 'tfMRI_GAMBLING_RL'
file = path_string.glob("tfMRI_GAMBLING_RL_Atlas_MSMAll.dtseries.nii")
try: 
    next(file)
    print("found")
except StopIteration:
    print("no file")
```

```python
with_gambling = []

for sub_path in hcp_subdirs:
    path_string  = sub_path / 'MNINonLinear' / 'Results' / 'tfMRI_GAMBLING_RL' 
    gen = path_string.glob("tfMRI_GAMBLING_RL_Atlas_MSMAll.dtseries.nii")
    try: 
        next(gen)
        patient = int(str(sub_path)[-6:])
        with_gambling.append(patient)
    except StopIteration:
        print("no file")
```
