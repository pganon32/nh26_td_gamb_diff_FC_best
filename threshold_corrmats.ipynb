```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
home = Path.home()
data_dir = Path(home / "shared" / "data" / "td-brain")
file_path = data_dir / "training_corrmats_all.npy"

# Load the array
data = np.load(file_path)
```

```python
np.shape(data)
```

```python
means = np.mean(data, axis=0)[1:]
np.shape(means)
```

```python
plt.plot(means[:200])
```

```python
# top 25% threshold (both neg and pos) 
upper_mean_thresh = np.sort(means)[int(0.875 * 64980)]
lower_mean_thresh = np.sort(means)[int(0.125 * 64980)]
```

```python
plt.plot(np.sort(means))
plt.axvline(x=0.875 * 64980, color='red', linestyle='--', linewidth=2)
plt.axvline(x=0.125 * 64980, color='red', linestyle='--', linewidth=2)
plt.axhline(y=upper_mean_thresh, color='r', linestyle='--', linewidth=2)
plt.axhline(y=lower_mean_thresh, color='r', linestyle='--', linewidth=2)
```
