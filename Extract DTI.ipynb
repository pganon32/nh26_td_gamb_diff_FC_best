```python
from pathlib import Path
from cloudpathlib import S3Path, S3Client
import pandas as pd

# Cache setup
cache_path = Path('/tmp/cache')
if not cache_path.exists():
    cache_path.mkdir()

# Client for the open-neurodata bucket (no auth needed)
client = S3Client(
    local_cache_dir=cache_path,
    no_sign_request=True)

hcp_derivs_path = S3Path(
    "s3://open-neurodata/rokem/hcp1200/afq",
    client=client)

# Client for the HCP bucket (needs auth) — to get the list of subjects
hcp_client = S3Client(
    local_cache_dir=cache_path,
    profile_name="hcp")

hcp_base_path = S3Path(
    "s3://hcp-openaccess/HCP_1200/",
    client=hcp_client)

# Get list of subject IDs
subject_ids = [p.name for p in hcp_base_path.iterdir()]
print(f"Found {len(subject_ids)} subjects")

#subject_ids = subject_ids[:5]  # test with just 5 subjects first  ← ADD IT HERE

# Loop through subjects, collect mean tract profiles
all_means = []

for sub_id in subject_ids:
    csv_path = (hcp_derivs_path / f"sub-{sub_id}" / "ses-01" /
                f"sub-{sub_id}_dwi_space-RASMM_model-CSD_desc-prob-afq_profiles.csv")
    try:
        if not csv_path.exists():
            print(f"Skipping {sub_id}: file not found")
            continue
        tract_profiles = pd.read_csv(csv_path)
        means = tract_profiles.groupby("tractID").mean(numeric_only=True)
        means['subjectID'] = sub_id
        all_means.append(means.reset_index())
    except Exception as e:
        print(f"Error with {sub_id}: {e}")
        continue

# Combine all subjects into one dataframe
final_df = pd.concat(all_means, ignore_index=True)

# Save to CSV
final_df.to_csv('all_subjects_tract_means.csv', index=False)
print("Saved to all_subjects_tract_means.csv")
```
