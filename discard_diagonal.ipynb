```python
from pathlib import Path
import numpy as np
from nilearn.connectome import vec_to_sym_matrix, sym_matrix_to_vec 
```

```python
data_path_win = Path("/home/jovyan/shared/data/td-brain") / "win_training_corrmats_all_072326.npy"
data_path_loss = Path("/home/jovyan/shared/data/td-brain") / "loss_training_corrmats_all_072326.npy"

corr_mat_win = np.load(data_path_win)
corr_mat_win_noID = corr_mat_win[:, 1:]
corr_mat_loss_noID = np.load(data_path_loss)
corr_mat_loss_noID = corr_mat_loss[:, 1:]
```

```python
n_nodes = 360

def drop_diagonal(vec_with_diag, n_nodes=360):
    matrices = vec_to_sym_matrix(vec_with_diag) 
    return sym_matrix_to_vec(matrices, discard_diagonal=True)

corr_mat_win_nodiag = drop_diagonal(corr_mat_win_noID, n_nodes)
corr_mat_loss_nodiag = drop_diagonal(corr_mat_loss_noID, n_nodes)

print(corr_mat_win_nodiag.shape)   # should be (n_subjects, 64620)
print(corr_mat_loss_nodiag.shape)  # should be (n_subjects, 64620)
```

```python
# extract IDs
subject_ids_win = corr_mat_win[:, 0]
subject_ids_loss = corr_mat_loss[:, 0]

# stitch the ID back as the first column
corr_mat_win_final = np.column_stack([subject_ids_win, corr_mat_win_nodiag])
corr_mat_loss_final = np.column_stack([subject_ids_loss, corr_mat_loss_nodiag])

print(corr_mat_win_final.shape)   # should be (692, 64621)  <- 64620 edges + 1 ID column
print(corr_mat_loss_final.shape)
```
